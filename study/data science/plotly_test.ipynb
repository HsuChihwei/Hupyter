```python
import plotly_express as px
gapminder = px.data.gapminder()
gapminder2007 = gapminder.query('year == 2007')
px.scatter(gapminder2007, x='gdpPercap', y='lifeExp')
gapminder.columns
gapminder.country.unique()
gapminder[gapminder.country=='Hong Kong, China']
# gapminder.continent.unique()
# gapminder[gapminder.continent=='Africa'].country.unique()
```

```python
# gapminder.query("country == 'China'")
px.scatter(gapminder[gapminder['country'].isin(['China', 'Hong Kong, China', 'Taiwan','Japan', 'Korea, Dem. Rep.','Korea, Rep.','France', 
                                                'Germany', 'United Kingdom', 'Egypt', 'South Africa','United States','Brazil','New Zealand','Australia'])],
           x='year', y='gdpPercap', color='country', size='pop', size_max=60, facet_col="continent",log_x=True)
# data(df), x-axis by first factor, y-axis by second factor, color by third factor, size by fourth factor, control the max size of fourth factor, columns by five factor, log the x-axis
```

```python
px.scatter(gapminder, x='gdpPercap', y='lifeExp', color='continent', size='pop', size_max=60, hover_name='country',animation_frame='year',animation_group='country',
           log_x=True,range_x=[100,100000], range_y=[20,90],labels={'pop': 'Population', 'gdpPercap':'GDP per Capita', 'lifeExp':'Life Expectancy'}
          )
# hover_name set focus point 
# animation_frame: dynamic changes by factor ( usually time type )
# animation_group
# log_x:  log the x-axis
# range_x: set x-axis value interval
# range_y: set y-axis value interval
# labels: set the key mapping of factors
```
