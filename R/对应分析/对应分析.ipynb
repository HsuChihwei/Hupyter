### 卡方独立性检验

```r
smoke = c(15,15)
nosmoke = c(5,65)
a = data.frame(smoke,nosmoke,row.names=c('ill', 'noill'))
a
```

```r
chisq.test(a)
```

* p值小于0.05拒绝原假设，所以，吸烟与得病是有关系的

### 对应分析

library(mass)

```r
h = c(121,188,112,86)
mh = c(57,105,65,60)
m = c(72,141,77,94)
ml = c(36,97,54,78)
l = c(21,71,54,71)
a = data.frame(h,mh,m,ml,l,row.names=c('好','轻微','中等','受损'))
a
```

```r
chisq.test(a)  # 卡方检验
```

* 通过卡方检验

```r
library(MASS)
```

```r
ca1 = corresp(a,nf=3) # nf提取公因子数
ca1
```

```r
summary(ca1)
```

```r
par(mar=c(4,4,3,1),cex=0.6)
```

```r
biplot(ca1)
abline(h=0,v=0,lty=2) #lty 线型
```
