### 随机生成三个簇点

```r
 c1<-cbind(rnorm(40,2,1),rnorm(40,2,3))
 c2<-cbind(rnorm(30,3,1),rnorm(30,20,1))
 c3<-cbind(rnorm(30,15,1),rnorm(30,25,1))  # 按列合并
```

```r
v<-rbind(c1,c2,c3)  # 按行合并
plot(v)
```

```r
dim(v)
```

```r
cl=kmeans(scale(v),3)
```

```r
cl
```

```r
plot(v, col=cl$cluster)
```

```r
# points(cl$centers,cex=1.2,col=1:2,pch=8)
points(cl$centers, col = 1:2, pch = 8, cex = 2)
```
