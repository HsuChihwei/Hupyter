```python
call_log_pd.call_duration.index.name = 'times'
call_log_pd.call_duration.rename('duration')
call_duration_data = call_log_pd.call_duration.astype(str).astype(int)
call_duration_data.plot()
```

```python
import matplotlib.pyplot as plt
call_rets = np.diff(np.log(call_duration_data))
plt.plot(call_rets)
```

```python
ax = plt.hist(call_rets)
```

```python
student = pd.DataFrame({ 
        'name': ['Smith', 'Brown', 'Phelps'],       
        'student_number': [17, 8, 666]
        })
student
```

```python
grade_report = pd.DataFrame({
        'student_number': [17, 17, 8, 8, 8, 8],
        'section_identifier': [112, 119, 85, 92, 102, 135],
        'grade': ['B', 'A', 'A', 'A', 'B', 'A']
        })
grade_report
```

```python
student.merge(grade_report, on='student_number')
```

```python
grade_report.merge(student, on='student_number')
```

```python
student.merge(grade_report, on='student_number', how='left')
```

```python
grade_report.merge(student, on='student_number', how='left')
```

```python
student.merge(grade_report, on='student_number', how='right')
```

```python
grade_report.merge(student, on='student_number', how='right')
```
