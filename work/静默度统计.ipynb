```python
BLANK_DEGREE_MAP = {
    'blank_times_count_30': ([1, 1, 2, float('inf')], [0,1,2,3], 0.3), #0.3
    'blank_times_count_90': ([1, 2, 3, float('inf')], [0,1,2,3], 0.25), #0.25
    'blank_times_count_180': ([1, 2, 6, float('inf')], [0,1,2,3], 0.15), #0.15
    'blank_times_count_max': ([72, 120, 360, float('inf')], [0,1,2,3], 0.3) #0.3
}

RESULT_MAPPING = ([0, 1, 2, 3], [0,1,2,3])

import bisect
def _cul_blank_degree(data):
    score_total = 0
    for k, v in BLANK_DEGREE_MAP.items():
        print(k,v,data[k])
        idx = bisect.bisect_right(v[0], int(float(data[k])))
        print idx
        score_total += v[1][idx]*v[2]
    idx_result = bisect.bisect_left(RESULT_MAPPING[0], score_total)
    return RESULT_MAPPING[1][idx_result]

data = {
    'blank_times_count_30': 3,
    'blank_times_count_90': 4,
    'blank_times_count_180': 6,
    'blank_times_count_max': 359,
}
_cul_blank_degree(data)
```
