```python
def mapping_keys(data, types=a):
    result = {}
    for key, value in data.items():
        if key in types:
            if isinstance(value, dict):
                if isinstance(types[key], dict):
                    result[key] = mapping_keys(value, types=types[key])
                else:
                    result[types[key]] = mapping_keys(value, types=types)
            elif isinstance(value, list):
                if isinstance(types[key], dict):
#                     result[key] = [mapping_keys(i, types=types[key]) if isinstance(i, dict) else i for i in value]
                    tmp = list()
                    for i in value:
                        if isinstance(i, dict):
                            tmp.append(mapping_keys(i, types=types[key]))
                        elif types.get(i):
                            tmp.append(types[i])
                        else:
                            tmp.append(i)
                    result[key] = tmp
                else:
#                     result[types[key]] = [mapping_keys(i, types=types) if isinstance(i, dict) else i for i in value]
                    tmp = list()
                    for i in value:
                        if isinstance(i, dict):
                            tmp.append(mapping_keys(i, types=types))
                        elif types.get(i):
                            tmp.append(types[i])
                        else:
                            tmp.append(i)
                    result[types[key]] = tmp
            else:
                result[types[key]] = value
        else:
            if isinstance(value, dict):
                if isinstance(types[key], dict):
                    result[key] = mapping_keys(value, types=types[key])
                else:
                    result[types[key]] = mapping_keys(value, types=types)
            else:
                result[key] = value
    return result
```
