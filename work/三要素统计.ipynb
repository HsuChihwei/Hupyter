```python
import json
with open("result.json",'w') as wp:
    with open("result.txt", 'r') as fp:
        code_list = list()
        result_dict = dict()
        for i in fp.readlines():
            tmp = i.split("\t")[0].replace('{"rescode":"','').replace('"}','').replace('{"rescode":','').replace('}','')
            if tmp == '\\N':
                tmp = -1
            tmp = int(tmp)
            tel = i.split("\t")[1].replace('phone":"','').replace('"\n','')
            if tmp not in result_dict:
                result_dict[tmp]=[]
            result_dict[tmp] = result_dict[tmp] + [tel,]
            if tmp not in code_list:
                code_list.append(int(tmp))
        code_set = set(code_list)
    json.dump(result_dict, wp, indent=4, sort_keys=True)

print code_set
# set([0, 1, 2, 3, 4, 5, 11, 12, 13, 14, 21, 22, 23, 24, -1])
```
