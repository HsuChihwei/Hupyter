### 按省份、性别分析占比

```python
prov_gender = df[["省级政区名称", "性别"]].dropna()
# .crosstab(行，列)用于针对字符串数据的透视（类似excel的数据透视）
pg = pd.crosstab(prov_gender["省级政区名称"], prov_gender["性别"])
pg.head()
```

```python
pg["女性占比"] = pg["女"] / (pg["男"] + pg["女"])
# .sort_values()排序，ascending = False表示降序
pg = pg.sort_values(by=["女性占比"], ascending=False)
pg.head()
```
