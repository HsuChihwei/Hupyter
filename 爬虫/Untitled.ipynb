```python
import requests
from lxml import etree
```

```python
def get_items(item):
    data = requests.get(f'http://www.jdcsww.com/qcgglist?ggxh={item}&ggpc=&zwpp=&clmc=&fdjxh=&qymc=&cph=&sbdm=&viewtype=1').text
    html = etree.HTML(data)
    pici = html.xpath('///*[@id="frmcontent"]/tbody/tr/td[2]/a/text()')
    date = html.xpath('///*[@id="frmcontent"]/tbody/tr/td[3]/a/text()')
    return pici if pici else '', date if date else ''
```

```python
# !pip install xlrd
import pandas as pd
```

```python
df['批次','发布时间'] = df['车辆型号'].map(get_items)
```

```python
av = pd.DataFrame((x for x in df[("批次", "发布时间")]),index=df.index,columns=['批次','发布时间'])
```

```python
qa = pd.concat( [df, av], axis=1 ) 
```
